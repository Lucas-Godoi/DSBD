# Treinamento de uma rede MLP para relação XOR - Tensorflow.keras

Este notebook mostra o uso do [API Keras](https://keras.io/api/models/) para criar e treinar uma rede Perceptrons Multicamadas (Multi-Layers Perceptrons - MLP) para a relação lógica XOR **e mostra os valores dos pesos sinápticos que são atribuídos à rede na sua inicialização e os valores de pesos finais (pós treinamento)**.

## Relação XOR (dados de entrada e saída)



```
import numpy as np
# Dados de entrada
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]],
            dtype=float)
# Saídas XOR
y_XOR = np.array([0, 1, 1, 0])
```

## Cria e treina uma rede MLP - configuração mínima para o XOR

Cria e treina rede uma rede MLP com dois neurônios na camada escondida (intermediária):

image.png



**Configuração mínima para o aprendizado da relação XOR.**



```
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import SGD
#create model
model = Sequential()

#add model layers
model.add(Dense(2, activation='sigmoid', input_shape=(2,)))
model.add(Dense(1))
```

```
model.summary()
```

## Valores iniciais dos pesos sinápticos

#### Camada oculta

```
weight1 = model.layers[0].get_weights()
weight1
```

#### Camada de saída

```
weight2 = model.layers[1].get_weights()
weight2
```

```
sgd = SGD(learning_rate=0.25)
model.compile(optimizer=sgd, loss='mean_squared_error')
```

```
model.fit(X, y_XOR,epochs=2000)
```

## Predição (teste da rede MLP treinada)


```
model.predict(X)
```

## Valores finais dos pesos (pós treinamento)

#### Camada oculta

```
weight1 = model.layers[0].get_weights()
weight1
```

#### Camada de saída

```
weight2 = model.layers[1].get_weights()
weight2
```
